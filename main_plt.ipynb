```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,3,7,8,1,3,9,11]

plt.plot(x,y, marker='o', color="Green")
plt.title("Example1")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]

plt.bar(x,y, color="Red")
plt.title("Example2")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]

plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red","green","blue","orange"])
plt.title="Example3"
plt.show()
```

```python
import numpy as np
x = np.random.rand(1000, 1)

plt.hist(x, bins=10)
plt.show()
```

```python
x = np.random.rand(64,64)

plt.imshow(x, cmap="hot")
plt.colorbar()
plt.show()
```
